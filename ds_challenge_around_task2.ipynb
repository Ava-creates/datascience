```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split


df = pd.read_csv('wine.csv')

df.describe()
df.drop_duplicates()
```

```python

df.head()
```

```python
df['quality']=df.quality.replace({"bad":0,"good":1})
df.head()
```

```python
y = df.quality.replace({"bad":0,"good":1})

```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def hypothesis(X, t):
    z = np.dot(t, X.T)
    return (sigmoid(z) - 0.0000001)


#X list of features
def cost(X, y, t):
    y1 = hypothesis(X, t)
    return -(1/len(X)) * np.sum(y*np.log(y1) + (1-y)*np.log(1-y1))


def gradient_descent(X, y, t, alpha, epochs):
    m =len(X)
    J = [cost(X, y, t)] 
    for i in range(0, epochs):
        h = hypothesis(X, t)
        for i in range(0, len(X.columns)):
            t[i] -= (alpha/m) * np.sum((h-y)*X.iloc[:, i])
        J.append(cost(X, y, t))
    return J, t

def predict(X, y, t, a, epochs):
    J, t = gradient_descent(X, y, t, a, epochs) 
    h = hypothesis(X, t)
    for i in range(len(h)):
        h[i]=1 if h[i]>=0.5 else 0
    y = list(y)
    acc = np.sum([y[i] == h[i] for i in range(len(y))])/len(y)
    return J, acc
```

```python
X = df.drop('quality',axis=1)
#y = df['quality']
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2,random_state=0)




theta = [0.5]*len(X_train.columns)
J, acc = predict(X_train, y_train, theta, 0.0001, 80000)

print(acc*100)
```

```python
theta = [0.5]*len(X_test.columns)
J, acc_test = predict(X_test, y_test, theta, 0.0001, 80000)
print(acc_test*100)

#accuracy of the test model 
```
